```python
import pandas as pd
```

```python
# Importar o arquivo csv
data = pd.read_csv('diabetes_dataset.csv')
data.head()
```

```python
# Limpar as linhas com dados faltantes
data = data.dropna().reset_index(drop=True)
data.head()
```

```python
# Importar bibliotecas necessárias para normalização
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np

# Definir as colunas que serão normalizadas (todas exceto Outcome)
colunas_para_normalizar = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
                           'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Separar features (X) e target (y)
X = data[colunas_para_normalizar]
y = data['Outcome']

print("Dados originais:")
print(X.head())
```

```python
# Normalização entre 0 e 1
scaler_minmax = MinMaxScaler()
X_normalized_minmax = scaler_minmax.fit_transform(X)
X_normalized_minmax_df = pd.DataFrame(X_normalized_minmax, columns=colunas_para_normalizar)

print("Normalização Min-Max (0 a 1):")
print(X_normalized_minmax_df.head())
```

```python
# Retorna a coluna de target ao dataframe e salva ele normalizado
X_normalized_minmax_df['Outcome'] = y
X_normalized_minmax_df.to_csv('diabetes_dataset.csv', index=False)
print(X_normalized_minmax_df.head())
```
